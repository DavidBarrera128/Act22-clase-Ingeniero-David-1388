```python
class Ingeniero:
    def __init__(self, nombre, especialidad, experiencia):
        self.nombre = nombre
        self.especialidad = especialidad
        self.experiencia = experiencia
        self.en_proyecto = False

    def asignar_proyecto(self, nombre_proyecto):
        self.en_proyecto = True
        print(f"\n[SISTEMA]: El Ing. {self.nombre} ha sido asignado con éxito.")
        print(f"Proyecto actual: {nombre_proyecto}.")

    def mostrar_perfil(self):
        print(f"\n--- FICHA DEL INGENIERO ---")
        print(f"Nombre: {self.nombre}")
        print(f"Especialidad: {self.especialidad}")
        print(f"Experiencia: {self.experiencia} años")
        print("-" * 25)

# --- Entrada de datos por el usuario ---

print("--- REGISTRO DE PERSONAL DE INGENIERÍA ---")

# Pedir datos para el Ingeniero 1
nom1 = input("Nombre del primer ingeniero: ")
esp1 = input("Especialidad: ")
exp1 = input("Años de experiencia: ")
proy1 = input("Nombre del proyecto a asignar: ")

ing1 = Ingeniero(nom1, esp1, exp1)

# Pedir datos para el Ingeniero 2
print("\n" + "."*30)
nom2 = input("Nombre del segundo ingeniero: ")
esp2 = input("Especialidad: ")
exp2 = input("Años de experiencia: ")
proy2 = input("Nombre del proyecto a asignar: ")

ing2 = Ingeniero(nom2, esp2, exp2)

# --- Ejecución de métodos ---

ing1.mostrar_perfil()
ing1.asignar_proyecto(proy1)

ing2.mostrar_perfil()
ing2.asignar_proyecto(proy2)

print("\n" + "="*40)
print("PROCESO DE CONTRATACIÓN Y ASIGNACIÓN COMPLETADO")
print("Creador del software por David Barrera NL = 8")
```
